```python
import re

with open("../data/HOSPITAL.md", "r", encoding="utf-8") as f:
    # Lấy nội dung văn bản từ file
    content = f.read() 
    
    pattern = r"(\d+)\s+(\d{2}/\d{2}/\d{4})\s*\n\s*(.+?)\s*\n\s*(https?://\S+)"
    
    # Sử dụng 'content' thay vì 'f'
    matches = re.findall(pattern, content)
    
    # Chuyển đổi thành list of tuples
    hospital_list = [[int(m[0]), m[1], m[2].strip(), m[3]] for m in matches]

print(f"Đã trích xuất được {len(hospital_list)} bệnh viện.")
```

```python
import os
from supabase import create_client, Client
from dotenv import load_dotenv
import random
load_dotenv()

supabase: Client = create_client(
    os.environ.get("SUPABASE_URL"),
    os.environ.get("SUPABASE_KEY")
)
```

```python
from datetime import datetime
data_to_upsert = [
    {"hospital_name": name, 
     "level": random.choice(["1", "2", "3"]), 
     "address": link,
     "created_at": datetime.strptime(date, "%d/%m/%Y").strftime("%Y-%m-%d")} 
    for code, date, name, link in hospital_list
]
response = supabase.table("hospitals").upsert(data_to_upsert).execute()
print("Hoàn thành cập nhật!")
```
