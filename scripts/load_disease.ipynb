```python
disease_data = []
with open("../data/DISEASE.md", "r", encoding='utf-8') as f:
    for line in f:
        tup = line.split(":")
        disease_data.append(tuple(tup)) if all([len(tup[0]) < 4, tup[0] not in ["I", "II", "III", "IV", "V", "VI", "VII", "VIII", 
                                                              "IX", "X", "XI", "XII", "XIII", "XIV", "XV", "XVI", 
                                                              "XVII", "XVIII", "XIX", "XX", "XXI", "XXII"]]
                                                              ) else None
```

```python
import os
from supabase import create_client, Client
from dotenv import load_dotenv

load_dotenv()

supabase: Client = create_client(
    os.environ.get("SUPABASE_URL"),
    os.environ.get("SUPABASE_KEY")
)
```

```python
data_to_upsert = [
    {"disease_code": code, "disease_name": name} 
    for code, name in disease_data
]
response = supabase.table("diseases_dictionary").upsert(data_to_upsert).execute()
print("Hoàn thành cập nhật!")
```
